# There are two ways to calling the model from Hugging Face
* downloading the model which is not relaibale way
* using **EndPointAPI** to call and use the model
* Using **Pipleine**


```python
from langchain_huggingface import HuggingFaceEndpoint
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
```

```python
Repo_id="mistral-community/pixtral-12b"
llm=HuggingFaceEndpoint(
    endpoint_url="https://api-inference.huggingface.co/models/"+Repo_id,
    huggingfacehub_api_token=sec_key,
    task="text-generation")
```

```python
question="what is machin learning"
template="""
Question:{question}
answer:here step by step
"""
prompt=PromptTemplate(template=template,input_variables=["question"])
```

```python
llm_chain=LLMChain(prompt=prompt,llm=llm)
```

##Pipeline method

```python
from langchain_huggingface import HuggingFacePipeline
from transformers import AutoTokenizer,AutoModelForCausalLM
```

```python
model_id="gpt2"
tokenizer=AutoTokenizer.from_pretrained(model_id)
model=AutoModelForCausalLM.from_pretrained(model_id)
```

```python
from transformers import pipeline
pipe=pipeline("text-generation",model=model,tokenizer=tokenizer)
```

```python
hf=HuggingFacePipeline(pipeline=pipe)
```

```python
hf.invoke("what is machin learning ")
```
